```python
# Read image
import cv2
import matplotlib.pyplot as plt
import numpy as np
im = cv2.imread("images/neg1.png")
# Apply negation
neg_img = 255 - im
# Display original and negative images
fig,ax = plt.subplots(1,2)
ax[0].imshow(im)
ax[1].imshow(neg_img)
ax[0].set_title("Original")
ax[1].set_title("Negated")

for a in ax.ravel():
    a.axis(False)
plt.tight_layout()
plt.show()
```

```python
# Read image
im = cv2.imread("images/log2.jpg")
im = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)

log_im = im.astype(np.float32)
log_trnx = np.log(1+im)
#Compute log transform image
c_log = 1 / np.log(1 + np.max(log_im))
#compute scale constant c_log
scaled_log_trnx  = c_log * log_trnx
# Apply scale contant to log_transformed image

#Display images
fig,ax = plt.subplots(1,2)
ax[0].imshow(im)
ax[1].imshow(scaled_log_trnx)
ax[0].set_title("Original")
ax[1].set_title("Log Transformed")

for a in ax.ravel():
    a.axis(False)
plt.tight_layout()
plt.show()
```

```python

#read image
im = cv2.imread("images/log2.jpg")
im = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)

#convert pixel values to range 0-1
im = im/255.0
# Apply Power-Law (Gamma) Transformation
gamma = 0.4
gamma_im = np.pow(im,gamma)
#Display images
fig,ax = plt.subplots(1,2)
ax[0].imshow(im)
ax[1].imshow(gamma_im)
ax[0].set_title("Original")
ax[1].set_title("Gamma Transformation")

for a in ax.ravel():
    a.axis(False)
plt.tight_layout()
plt.show()
```

<h3>Image Transformation operations</h3>
Image transformation functions operate directly on the pixels to modify the visual appearance of images.
<h4>Motivation</h4>
Digital images often suffer from poor contrast, non-uniform illumination, due to sensor limitations or acquisition conditions. Simple intensity transformations allow us to re-map pixel values to enhance certain regions of interest without altering the spatial structure of the image and hence produce a more useful representation of the image

<h3>Negative Transformation</h3>
The negative of an image is obtained by inverting its intensity values i.e. black becomes white and white becomes black. This transformation is useful when image details are embedded in darker regions, as it can enhance visibility by reversing intensity relationships.

<code>negative_img = 255-img</code>


<h3>Logarithmic Transformation</h3>
Logarithmic transformations expand the values of dark pixels while compressing higher intensity values. Preferable when you want to brighten darker regions of the image without saturating brighter regions.
<p><i>Change each pixel’s brightness in a non-linear way so that dark pixels change a lot, and bright pixels change only a little</i>
<h5>Algorithm</h5>
<ul>
    For each pixel
    <li>Add 1 to its intensity (to avoid log(0) issues)</li>
    <li>Apply logarithm</li>
    <li>Scale the result so that the brightest pixel does not exceed the allowed range
    <code>scale_factor = 1/log(1+max_pixel_value)</code>
    </li>
</ul>    


<h3>Power-Law (Gamma) Transformation</h3>
Power-law transformations provide flexible control over image brightness and contrast through the gamma parameter. By adjusting gamma, images can be made brighter or darker in a controlled manner.
<p>
    Expects pixel values to be in the range 0-1
</p>
<code>
gamma_image = image ^ gamma
</code>
<br/>
if gamma < 1 ==> results in brighter image
<br/>
if gamma > 1 ==> results in darker image
