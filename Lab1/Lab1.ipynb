<h2>Lab 1</h2>
Write a Program to read a color digital image and perform the following:
<ol>
    <li>Split and display its RGB channels. Apply zeros to other channels, display RGB
channels again, and observe the differences.</li>
    <li>Convert it to grayscale and binary image. Display all the images one by one
along with its histogram.</li>
</ol>

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
# 1. Load the image
im = cv2.imread('bird.jpeg')
# 2 Convert BGR2RGB
im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
# 3 Split image into 3 channels R, G & B
R,G,B = cv2.split(im_rgb)
fig, ax = plt.subplots(1,3)
ax[0].axis('off')
ax[0].set_title('Red component')
ax[0].imshow(R, cmap='gray')
ax[1].axis('off')
ax[1].set_title('Green component')
ax[1].imshow(G, cmap='gray')
ax[2].axis('off')
ax[2].set_title('Blue component')
ax[2].imshow(B, cmap='gray')

# 4. Create a zero matrix: zeroes with the same size as R
z = np.zeros_like(R)
# 5. Create 3 new images by merging the following channels (use cv2.merge)
# only_R = R + zeroes + zeroes
only_R = cv2.merge([R,z,z])
# only_G = zeroes + G + zeroes
only_G = cv2.merge([z,G,z])
# only_B = zeroes + zeroes + B
only_B = cv2.merge([z,z,B])

#6. Display the final image as shown in the reference output figure above
fig, ax = plt.subplots(1,3)
ax[0].axis('off')
ax[0].set_title('Red with zeros')
ax[0].imshow(only_R)
ax[1].axis('off')
ax[1].set_title('Green with zeros')
ax[1].imshow(only_G)
ax[2].axis('off')
ax[2].set_title('Blue with zeros')
ax[2].imshow(only_B)
```

```python
def display_image_and_histogram(image, plot_title, cmap=None):
  fig, ax = plt.subplots(1, 2, figsize=(8, 2))
# Show Image with relevant title at ax[0]
  ax[0].imshow(image, cmap=cmap)
  ax[0].set_title(plot_title)
  ax[0].axis('off')
#show histogram with relevant title at ax[1]
  ax[1].hist(image.ravel())

# 1. Load image
im = cv2.imread('bird.jpeg')
# 2. Create a grayscale image 'gray' using relevant formula
B,G,R = cv2.split(im)
gray = 0.299*R+0.587*G+0.114*B
# 3. Ask user to enter threshold 'th'
th = int(input("Enter binary threshold value: "))
# 4. Create a binary image using the threshold value
binary = gray.copy()
binary[binary>=th]=255
binary[binary<th]=0
# Display all images with histograms
display_image_and_histogram(im, "Original")
display_image_and_histogram(gray, "Grayscale", cmap="gray")
display_image_and_histogram(binary, "Binary Image", cmap="gray")
```

<h2>Grayscale Image and Binary Image</h2>
<h3>Grayscale Images:</h3>
<ul>
    <li>A grayscale image is one in which the value of each pixel represents only the intensity of light, with no color information. Typically, the intensity values range from 0 (black) to 255 (white). Grayscale conversion simplifies image analysis by reducing the complexity of the color data to a single channel.
</li>
    <li><b>Applications:</b> Grayscale images are used in many applications where color is not essential, such as edge detection, texture analysis, and medical imaging.</li>
</ul>

The formula to convert RGB to grayscale is
<p>
  Grayscale (Y) = (0.299 &times; R) + (0.587 &times; G) + (0.114 &times; B)
</p>


<h3>Binary Image</h3>
<ul>
    <li>A binary image is one where each pixel is either black (0) or white (255). This is obtained by thresholding a grayscale image, where pixels above a certain intensity value are set to white (255), and those below are set to black (0).</li>
    <li>Binary images are useful for tasks that require clear distinction between objects and the background.</li>
    <li><b>Applications:</b> Binary images are used in object detection, shape analysis, and image segmentation.
</li>
</ul>

<h3>Histogram of an image</h3>
Histogram refers to the count of pixels of a particular value in the image. That is, how many pixels with value 0, how many with value 1, etc.

Matplotlib provides a function hist() to create hist plot. hist() takes as input a one-dimensional array and generates a plot representing the frequency of different values in that array. Images contain 2 or 3 dimensions which cannot be directly used in hist() method. So we use image.ravel() to flatten the 2D/3D image to a one dimensional image
Syntax:
<code>plt.hist(image.ravel())</code>
