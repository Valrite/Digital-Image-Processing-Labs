```python
# Import the required libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np
# 1. Load the image
im = cv2.imread('pet.jpg')
# 2 Convert BGR2RGB
im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
# 3 Split image into 3 channels R, G & B
R,G,B = cv2.split(im_rgb)
fig, ax = plt.subplots(1,3)
ax[0].axis('off')
ax[0].set_title('Red component')
ax[0].imshow(R, cmap='gray')
ax[1].axis('off')
ax[1].set_title('Green component')
ax[1].imshow(G, cmap='gray')
ax[2].axis('off')
ax[2].set_title('Blue component')
ax[2].imshow(B, cmap='gray')

# 4. Create a zero matrix: zeroes with the same size as R
z = np.zeros_like(R)
# 5. Create 3 new images by merging the following channels (use cv2.merge)
# only_R = R + zeroes + zeroes
only_R = cv2.merge([R,z,z])
# only_G = zeroes + G + zeroes
only_G = cv2.merge([z,G,z])
# only_B = zeroes + zeroes + B
only_B = cv2.merge([z,z,B])

#6. Display the final image as shown in the reference output figure above
fig, ax = plt.subplots(1,3)
for a in ax.ravel():
  a.axis("off")
ax[0].set_title('Red with zeros')
ax[0].imshow(only_R)
ax[1].set_title('Green with zeros')
ax[1].imshow(only_G)
ax[2].set_title('Blue with zeros')
ax[2].imshow(only_B)
```

```python
# Function to display an image and its histogram
def display_image_and_histogram(image, plot_title, cmap=None):
  fig, ax = plt.subplots(1, 2, figsize=(8, 2))
# Show Image with relevant title at ax[0]
  ax[0].imshow(image, cmap=cmap)
  ax[0].set_title(plot_title)
  ax[0].axis('off')
#show histogram with relevant title at ax[1]
  ax[1].hist(image.ravel())

# 1. Load image
im = cv2.imread('pet.jpg')
# 2. Create a grayscale image 'gray' using relevant formula
B,G,R = cv2.split(im)
gray = 0.299*R+0.587*G+0.114*B
# 3. Ask user to enter threshold 'th'
th = int(input("Enter binary threshold value: "))
# 4. Create a binary image using the threshold value
binary = gray.copy()
binary[binary>=th]=255
binary[binary<th]=0
# Display all images with histograms
display_image_and_histogram(im, "Original")
display_image_and_histogram(gray, "Grayscale", cmap="gray")
display_image_and_histogram(binary, "Binary Image", cmap="gray")
```
